```python
import networkx as nx
import matplotlib.pyplot as plt

from puzzle14 import inp1, inp2, inp3, inp4, inp5, parse_input_with_weights

def draw_graph(reactions):
    G = nx.DiGraph()
    for qty, name in reactions.keys():
        G.add_node(name)
    for key in reactions.keys():
        qty, name = key
        for reactant in reactions[key]:
            r_qty, r_name = reactant
            G.add_edge(r_name, name)

    fig, ax = plt.subplots(figsize=(7, 7))
    nx.draw_circular(G, with_labels=True, font_weight='bold')
```

```python
draw_graph(parse_input_with_weights(inp1))
```
