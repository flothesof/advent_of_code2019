# Running the program 

```python
from puzzle17 import prints2ascii
```

```python
# message: encoded in ASCII
# 1. main routine (A, B, C)
# 2. functions A, B, C separated by newlines (code 10)
# 3. y/n for video feed

# main
main = 'ABBCCABBCA'

# feed
feed = 'n'

message = []
message.extend([m for m in ",".join(main)] + ['\n'])
assert len(A) < 20
message.extend([inst for inst in A] + ['\n'])
message.extend([inst for inst in B] + ['\n'])
message.extend([inst for inst in C] + ['\n'])
message.extend([feed, '\n'])
message_ascii = [ord(val) for val in message]

# load the new program
program[0] = 2
computer = IntCodeComputer(program, resume=False)
prog_state, prints, status = computer.run(input_values=message_ascii)
ascii_prints = prints2ascii(prints, prune=False)
[print(row) for row in ascii_prints];
```
